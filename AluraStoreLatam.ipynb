```python
from itertools import combinations


# Convertir los índices de las Series en conjuntos
sets_tiendas = [
    set(top_10_productos_tienda1.index),
    set(top_10_productos_tienda2.index),
    set(top_10_productos_tienda3.index),
    set(top_10_productos_tienda4.index)
]

# Encontrar las coincidencias entre dos o más tiendas
for i in range(2, len(sets_tiendas) + 1):  # Iterar sobre combinaciones de 2, 3 y 4 tiendas
    for combinacion_tiendas in combinations(sets_tiendas, i):
        coincidencias = set.intersection(*combinacion_tiendas)
        if coincidencias:
            nombres_tiendas = [f"Tienda {j + 1}" for j in range(len(sets_tiendas)) if sets_tiendas[j] in combinacion_tiendas]
            print(f"Coincidencias entre {', '.join(nombres_tiendas)}:")
            for producto in coincidencias:
                print(f"- {producto}")
            print()
```

```python
# Contar las coincidencias de cada tienda
coincidencias_por_tienda = {f"Tienda {i + 1}": 0 for i in range(len(sets_tiendas))}

for i in range(2, len(sets_tiendas) + 1):
    for combinacion_tiendas in combinations(sets_tiendas, i):
        coincidencias = set.intersection(*combinacion_tiendas)
        if coincidencias:
            for j in range(len(sets_tiendas)):
                if sets_tiendas[j] in combinacion_tiendas:
                    coincidencias_por_tienda[f"Tienda {j + 1}"] += len(coincidencias)

# Encontrar la tienda con más coincidencias
tienda_mas_coincidencias = max(coincidencias_por_tienda, key=coincidencias_por_tienda.get)

print(f"La tienda con más coincidencias es: {tienda_mas_coincidencias} "
      f"con {coincidencias_por_tienda[tienda_mas_coincidencias]} coincidencias.")
```

```python
import plotly.express as px
import pandas as pd

# Combinar los DataFrames de las 4 tiendas en uno solo
all_tiendas = pd.concat([tienda1, tienda2, tienda3, tienda4], keys=['Tienda 1', 'Tienda 2', 'Tienda 3', 'Tienda 4'])
all_tiendas.reset_index(level=0, inplace=True)
all_tiendas.rename(columns={'level_0': 'Tienda'}, inplace=True)

# Agrupar por lugar de compra y tienda para obtener la cantidad de pedidos
lugares_compra = all_tiendas.groupby(['Lugar de Compra', 'Tienda'])['Producto'].count().reset_index()
lugares_compra.rename(columns={'Producto': 'Cantidad de Pedidos'}, inplace=True)

# Obtener coordenadas únicas para cada lugar de compra
coordenadas_lugares = all_tiendas.groupby('Lugar de Compra')[['lat', 'lon']].first().reset_index()

# Agregar las columnas 'lat' y 'lon' al DataFrame lugares_compra usando merge
lugares_compra = pd.merge(lugares_compra, coordenadas_lugares, on='Lugar de Compra', how='left')

# Crear el mapa con plotly.express, agregando interactividad
fig = px.scatter_mapbox(
    lugares_compra,
    lat="lat",
    lon="lon",
    color="Tienda",
    size="Cantidad de Pedidos",
    size_max=30,
    zoom=4,
    mapbox_style="carto-positron",
    hover_name="Lugar de Compra",
    hover_data={"Tienda": True, "Cantidad de Pedidos": True},
    center={"lat": 4.5709, "lon": -74.2973},
    category_orders={"Tienda": ["Tienda 1", "Tienda 2", "Tienda 3", "Tienda 4"]},
)

# Actualizar las trazas para la interactividad
fig.update_traces(
    visible=True,
    opacity=1,
    selector=dict(type='scattermapbox'),
)

# Configurar la interactividad de la leyenda
for i, tienda in enumerate(fig.data):
    fig.data[i].visible = True if i == 0 else 'legendonly'
    fig.data[i].opacity = 1 if i == 0 else 0.5

fig.update_traces(below="''")

# Agregar control de zoom
fig.update_layout(uirevision='foo')  # 'foo' puede ser cualquier valor

# Agregar título y aclaración
fig.update_layout(
    title_text="Destino de los Pedidos",
    title_x=0.5,  # Centrar el título

    margin=dict(l=250, r=50, t=50, b=50)  # Ajusta los márgenes según tus necesidades
)
fig.show()
```

### **Resumen Ejecutivo**

Este informe presenta un análisis de las ventas, costos de envío, valor neto de ventas y calificación promedio de cuatro tiendas. Se ha encontrado que la tienda 4 tiene las ventas netas más bajas y sus productos más vendidos podrían ser suplidos por las demás tiendas. Además, su calificación promedio, aunque no es la más baja, no justifica su continuidad en comparación con las demás. Por lo tanto, se considera la posibilidad de cerrar la tienda 4 para optimizar las operaciones.


### **Análisis de Ventas**

* La tienda 1 presenta las mayores ventas, mientras que la tienda 4 tiene las menores ventas. Sin embargo, la diferencia entre la tienda con mayores ventas (tienda 1) y la tienda con menores ventas (tienda 4) es de aproximadamente un 20%, lo cual no es una diferencia significativa.


### **Análisis de Costos de Envío**

* La tienda 1 tiene los costos de envío más altos, mientras que la tienda 4 tiene los costos más bajos. La diferencia entre estos dos extremos es de alrededor de un 10%.


### **Valor Neto de Ventas**

* El valor neto de ventas más alto lo tiene la tienda 1, con aproximadamente 1.1 mil millones de pesos. La tienda 4 tiene el valor neto de ventas más bajo, con alrededor de 980 millones de pesos.
